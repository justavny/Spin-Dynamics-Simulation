# Single Spin Dynamics

### 1. Introduction

A spin- ${\frac{1}{2}}$ particle is the simplest quantum mechanical system and can be represented using a single qubit.

The quantum state of a spin can be written as ${|\psi\rangle = \alpha |+\rangle + \beta |-\rangle}$, where ${|\alpha|^2 + |\beta|^2 = 1 }$ and ${|+\rangle}$ and ${|- \rangle}$ are the eigenstates of ${\hat{S}^z}$ with eigenvalues of ${\pm \frac{\hbar}{2}}$.   

The state evolves according to the Schrödinger equation

$${
i\frac{d}{dt}|\psi(t)\rangle = H|\psi(t)\rangle
}$$

where \(H\) is the Hamiltonian of the system.

#### Generating an arbitary state
Quantum hardware initializes each qubit in the state ${|+\rangle \equiv |0\rangle}$.  

An arbitrary spin state ${|\psi(\theta,\phi)\rangle = \cos\frac{\theta}{2}|+\rangle + e^{i\phi}\sin\frac{\theta}{2}|-\rangle}$ can be prepared from ${|+\rangle}$ using a single-qubit unitary operation. 
In Qiskit, this is implemented using the three-parameter gate  

$${
U_3(\theta,\phi,\lambda)
=
\begin{pmatrix}
\cos\frac{\theta}{2}
&
-e^{i\lambda}\sin\frac{\theta}{2}
\\[6pt]
e^{i\phi}\sin\frac{\theta}{2}
&
e^{i(\phi+\lambda)}
\cos\frac{\theta}{2}
\end{pmatrix}
}$$

Setting ${\lambda=0}$ generates

$${
|\psi(\theta,\phi)\rangle
=
U_3(\theta,\phi,0)
|+\rangle
}$$



### 2. Hamiltonian and Time Evolution

Consider a spin placed in a magnetic field directed along the x-axis. The Hamiltonian is ${H = h_x \hat S^x}$

where

$${
\hat S^x =
\begin{pmatrix}
0 & 1\\
1 & 0
\end{pmatrix}
}$$

which is also the Pauli-X matrix (considering ${\hbar}$ = 1).

For a time-independent Hamiltonian, the solution of the Schrödinger equation is

$${
|\psi(t)\rangle
=
e^{-iHt}
|\psi(0)\rangle
}$$

The operator ${ U(t)=e^{-iHt} }$ is called the **time-evolution operator**. It completely determines how the quantum state changes with time.


### 3. Unitary Operator for an Arbitrary Quantum State

Substituting the Hamiltonian,

$${
U(t) = e^{-ih_x t \hat S^x} = e^{-i(h_x t/2)\sigma_x}
}$$

Using ${\sigma_x^2=I}$ the exponential simplifies to

$${
U(t) = \cos\!\left(\frac{h_x t}{2}\right)I - i\sin\!\left(\frac{h_x t}{2}\right)\sigma_x
}$$  
  

### 4. Rotations and Why We Need Them

Since the computational basis corresponds to the eigenbasis of the spin operator $ S^z $, a direct measurement on the quantum computer determines only the spin projection along the z-axis. To measure spin projections along other directions, the state must first be rotated so that the desired measurement axis is mapped onto the z-axis. A standard measurement is then performed in the computational basis.

$${
R_x(\theta)
=
\begin{pmatrix}
1 & 0 & 0\\
0 & \cos\theta & -\sin\theta\\
0 & \sin\theta & \cos\theta
\end{pmatrix}
}$$  

is the Rotation matrix that rotates a 3D vector about the x-axis by $\theta$.  
  
A spin-${\frac{1}{2}}$ state, however, is represented by a two-component state vector on the Bloch sphere rather than a vector in real space. Consequently, rotations are generated by the spin operators.

A rotation of a quantum state by an angle ${\theta}$ about the axis ${\alpha=(x,y,z)}$ is given by

$${
\hat R_\alpha(\theta)
=
e^{-i\theta \hat\sigma^\alpha/2}
}$$

where ${\hat\sigma^\alpha}$ denotes the corresponding Pauli matrix.

Using the identity ${(\hat\sigma^\alpha)^2 = I}$

the rotation operator can be written as 
$${
\hat R_\alpha(\theta)
=
I\cos\frac{\theta}{2}
-
i\hat\sigma^\alpha
\sin\frac{\theta}{2}
}$$

Thus, a rotation of a spin state corresponds to a unitary transformation of the qubit state. In Qiskit, such operations are implemented through the rotation gates

$${
R_x(\theta),\qquad
R_y(\theta),\qquad
R_z(\theta)
}$$

These rotations are particularly useful for changing the measurement basis. Since quantum hardware performs measurements in the computational basis, corresponding to the eigenstates of $S^z$, suitable rotations allow measurements of spin projections along the $x$ and $y$ directions as well.

### Key Idea

The dynamics of a single spin in a magnetic field can be mapped directly onto the dynamics of a single qubit. The Hamiltonian generates a unitary evolution operator, which is equivalent to a qubit rotation gate. This allows spin dynamics to be simulated efficiently on quantum hardware.
